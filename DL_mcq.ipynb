# SingleRank Extractor
Extracting keywords which are Proper Nouns from summarized text using SingleRank algorithm

```python
pos = {'PROPN'}
extractor = pke.unsupervised.SingleRank()
extractor.load_document(input=summarized_text,stoplist=stoplist,language='en',normalization=None)
```

Single rank algorithm is a graph based unsupervised algorithm.It creates a weighted graph where nodes are words.A edge will be established only if two nodes exist within window size here the window size is 10.Here the concept is that each node contributes a vote to it's connected node.For a node all votes will be aggregated from it's connected nodes iteratively using google's pagerank algorithm.Weights of edges are assigned based on co-occurence of two nodes.Here the low scored nodes wont be filtered.The scores will be calculated based on it's constituent node scores.Here we are returning top 10 keyphrases.

```python
extractor.candidate_selection(pos=pos)
extractor.candidate_weighting(window=10,pos=pos)
keyphrases = extractor.get_n_best(n=10)
```

# MultipartiteRank Extractor
SingleRank is suitable for small or medium sized documents.On long documents singlerank fail to identify information about coocurence of words.It assumes keyphrases occur at beginning of document.So,it weighs nodes which favors words appear at beginning.Edges will be weighted based on coocurrence count of nodes attached to edge.The nodes will be weighted inverse of position in the document.In MultipartiteRank Keyphrases act as nodes and will be connected if key phrases are from different topics to form directed graph.

```python
import string
extractor1 = pke.unsupervised.MultipartiteRank()
stoplist = list(string.punctuation)
stoplist += pke.lang.stopwords.get('en')
extractor1.load_document(input=summarized_text,stoplist=stoplist)
```

```python
pos = {'PROPN'}
extractor1.candidate_selection(pos=pos)
extractor1.candidate_weighting(alpha=1.1,threshold=0.74,method='average')
keyphrases = extractor1.get_n_best(n=10)
```

```python
def get_distractors_wordnet(syn,word):
    distractors=[]
    word= word.lower()
    orig_word = word
    if len(word.split())>0:
        word = word.replace(" ","_")
    hypernym = syn.hypernyms()
    if len(hypernym) == 0: 
        return distractors
    for item in hypernym[0].hyponyms():
        name = item.lemmas()[0].name()
        #print ("name ",name, " word",orig_word)
        if name == orig_word:
            continue
        name = name.replace("_"," ")
        name = " ".join(w.capitalize() for w in name.split())
        if name is not None and name not in distractors:
            distractors.append(name)
    return distractors
```

# Extracting senses
Extracting the differents senses of the word and extracting the root word with the high similarity

```python
def get_wordsense(sent,word):
    word= word.lower()
    
    if len(word.split())>0:
        word = word.replace(" ","_")
    
    
    synsets = wn.synsets(word,'n')
    if synsets:
        wup = max_similarity(sent, word, 'wup', pos='n')
        adapted_lesk_output =  adapted_lesk(sent, word, pos='n')
        lowest_index = min (synsets.index(wup),synsets.index(adapted_lesk_output))
        return synsets[lowest_index]
    else:
        return None
```

# Conceptnet 
Conceptnet doesn't have any ambiguity it's straightforward in finding word irrespective of sense.For a particular word it always returns certain same set of words.

```python
def get_distractors_conceptnet(word):
    word = word.lower()
    original_word= word
    if (len(word.split())>0):
        word = word.replace(" ","_")
    distractor_list = [] 
    url = "http://api.conceptnet.io/query?node=/c/en/%s/n&rel=/r/PartOf&start=/c/en/%s&limit=5"%(word,word)
    obj = requests.get(url).json()

    for edge in obj['edges']:
        link = edge['end']['term'] 

        url2 = "http://api.conceptnet.io/query?node=%s&rel=/r/PartOf&end=%s&limit=10"%(link,link)
        obj2 = requests.get(url2).json()
        for edge in obj2['edges']:
            word2 = edge['start']['label']
            if word2 not in distractor_list and original_word.lower() not in word2.lower():
                distractor_list.append(word2)
                   
    return distractor_list
```

# Conceptnet Usage
Sometime every word not available in wordnet then we use conceptnet

```python
for keyword in keyword_sentences:
    wordsense = get_wordsense(keyword_sentences[keyword][0],keyword)
    if wordsense:
        distractors = get_distractors_wordnet(wordsense,keyword)
        if len(distractors) ==0:
            distractors = get_distractors_conceptnet(keyword)
        if len(distractors) != 0:
            key_distractor_list[keyword] = distractors
    else:
        
        distractors = get_distractors_conceptnet(keyword)
        if len(distractors) != 0:
            key_distractor_list[keyword] = distractors
```
